# VIOLENCIA DE GENERO BUCARAMANGA

```python
violencia=pd.read_csv("./violencia.csv")
print(violencia.columns)
violencia
```

```python
columnas=['Municipio', 'año', 'Grupo edad',
       'Ciclo de vida', 'sexo_',
       'Tipo de Seguridad Social',
       'naturaleza', 'def_naturaleza', 'actividad', 'nom_actividad',
       'edad_agre', 'sexo_agre', 'parentezco_vict','MES']

violencia=violencia[columnas]

violencia['sexo_'].replace({'Masculino':'1'},inplace=True)
violencia['sexo_'].replace({'Femenino':'2'},inplace=True)
violencia['sexo_']=pd.to_numeric(violencia['sexo_'])
violencia['Tipo de Seguridad Social'].replace({'No asegurado':'0'},inplace=True)
violencia['Tipo de Seguridad Social'].replace({'Contributivo':'1'},inplace=True)
violencia['Tipo de Seguridad Social'].replace({'Subsidiado':'2'},inplace=True)
violencia['Tipo de Seguridad Social'].replace({'Especial':'3'},inplace=True)
violencia['Tipo de Seguridad Social'].replace({'Excepción':'4'},inplace=True)
violencia['Tipo de Seguridad Social'].replace({'Indeterminado':'5'},inplace=True)
violencia['Tipo de Seguridad Social']=pd.to_numeric(violencia['Tipo de Seguridad Social'])
violencia['sexo_agre'].replace({'Masculino':'1'},inplace=True)
violencia['sexo_agre'].replace({'MASCULINO':'1'},inplace=True)
violencia['sexo_agre'].replace({'FEMENINO':'2'},inplace=True)
violencia['sexo_agre'].replace({'Femenino':'2'},inplace=True)
violencia['sexo_agre'].replace({'SIN DEFINIR':'0'},inplace=True)
violencia['sexo_agre']=pd.to_numeric(violencia['sexo_agre'])
violencia['MES'].replace({'enero':'1','febrero':'1','marzo':'1','abril':'1','marzo':'1','mayo':'1','junio':'1'},inplace=True)
violencia['MES'].replace({'julio':'2','agosto':'2','septiembre':'2','octubre':'2','noviembre':'2','diciembre':'2'},inplace=True)
violencia['MES']=pd.to_numeric(violencia['MES'])
violencia['Grupo edad'].replace({'<1':'0','1 a 5':'1','6 a 11':'6','12 a 17':'12','18 a 28':'18','29 a 59':'29','60 y mas':'60'},inplace=True)
violencia['Grupo edad']=pd.to_numeric(violencia['Grupo edad'])

violencia
```

```python
cs=['año', 'sexo_', 'Tipo de Seguridad Social',
       'naturaleza', 'actividad',
       'edad_agre', 'Grupo edad', 'sexo_agre', 'MES']
for c in cs:
    columna=c
    plt.hist(violencia[columna]) #1:H, 2:M
    plt.xlabel(columna)
    plt.show()
```

```python
plt.scatter(violencia['año'],violencia['def_naturaleza'],alpha=0.1,s=100) #1:H, 2:M
```

```python
i20=violencia['edad_agre']>0
violenciap=violencia[i20]
plt.scatter(violenciap['edad_agre'],violenciap['def_naturaleza'],alpha=0.1,s=100) #1:H, 2:M
plt.show()
plt.scatter(violenciap['Grupo edad'],violenciap['def_naturaleza'],alpha=0.1,s=100) #1:H, 2:M
plt.show()
fig,ax = plt.subplots(figsize=(5,10))
plt.scatter(violencia['def_naturaleza'],violencia['parentezco_vict'],alpha=0.1,s=100) #1:H, 2:M
ax.tick_params(axis='x', labelrotation=90)
plt.setp(ax.get_xticklabels(), rotation=45, ha='right')
plt.show()
fig,ax = plt.subplots(figsize=(5,5))
plt.scatter(violencia['def_naturaleza'],violencia['nom_actividad'],alpha=0.1,s=100) #1:H, 2:M
plt.setp(ax.get_xticklabels(), rotation=45, ha='right')
plt.show()
plt.scatter(violenciap['edad_agre'],violenciap['sexo_agre'],alpha=0.1,s=100) #1:H, 2:M
plt.show()
```
