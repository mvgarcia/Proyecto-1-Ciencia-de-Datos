```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sklearn.linear_model
```

```python
inscritos2020=pd.read_csv("./matriculados2020.csv")
inscritos2019=pd.read_csv("./matriculados2019.csv")
inscritos2018=pd.read_csv("./matriculados2018.csv")
inscritos2017=pd.read_csv("./matriculados2017.csv")

columns=['CÓDIGO DE LA INSTITUCIÓN', 'PRINCIPAL O SECCIONAL',
         'CARACTER IES','MUNICIPIO DE OFERTA DEL PROGRAMA',
       'INSTITUCIÓN DE EDUCACIÓN SUPERIOR (IES)',
       'ID SECTOR IES', 'SECTOR IES',
       'CÓDIGO SNIES DEL PROGRAMA', 'PROGRAMA ACADÉMICO', 'ID NIVEL ACADÉMICO',
       'NIVEL ACADÉMICO', 'ID NIVEL DE FORMACIÓN', 'NIVEL DE FORMACIÓN',
       'ID METODOLOGÍA', 'METODOLOGÍA', 'ID ÁREA DE CONOCIMIENTO',
       'ÁREA DE CONOCIMIENTO',
       'ID SEXO', 'SEXO', 'AÑO', 'SEMESTRE', 'MATRICULADOS']

inscritos2020=inscritos2020[columns]
inscritos2019=inscritos2019[columns]
inscritos2018=inscritos2018[columns]
inscritos2017=inscritos2017[columns]

inscritos2020 = pd.concat([inscritos2020, inscritos2019])
inscritos2020 = pd.concat([inscritos2020, inscritos2018])
inscritos2020 = pd.concat([inscritos2020, inscritos2017])
print(inscritos2020.columns)
inscritos2020
```

```python
i20=inscritos2020['PRINCIPAL O SECCIONAL'].str.lower()=='Principal'.lower()
inscritos2020=inscritos2020[i20]

i20=inscritos2020['CARACTER IES'].str.lower()=='Universidad'.lower()
inscritos2020=inscritos2020[i20]

i20=inscritos2020['MUNICIPIO DE OFERTA DEL PROGRAMA'].str.lower()=='Bucaramanga'.lower()
inscritos2020=inscritos2020[i20]

i20=(inscritos2020['NIVEL DE FORMACIÓN']=='Universitaria')+(inscritos2020['NIVEL DE FORMACIÓN']=='Maestría')+(inscritos2020['NIVEL DE FORMACIÓN']=='Doctorado')
inscritos2020=inscritos2020[i20]
inscritos2020
```

```python
columna='AÑO'
inscritos2020[columna]=pd.to_numeric(inscritos2020[columna])
plt.hist(inscritos2020[columna]) #1:H, 2:M
plt.xlabel(columna)
```

```python
plt.scatter(inscritos2020['MATRICULADOS'],inscritos2020['NIVEL DE FORMACIÓN'],alpha=0.005,s=100)
```

```python
#columna='ID SEXO'
columna='SEMESTRE'
inscritos2020[columna]=pd.to_numeric(inscritos2020[columna])
plt.hist(inscritos2020[columna]) #1:H, 2:M
plt.xlabel(columna)
```

```python
#columna='SEMESTRE'
ii_1=inscritos2020[columna]==1 #H
ii_2=inscritos2020[columna]==2 #M

copia=inscritos2020['MATRICULADOS']

n1=np.count_nonzero(ii_1)
n2=np.count_nonzero(ii_2)
print(n1,n2)

inscritos_1=copia[ii_1]
inscritos_2=copia[ii_2]

suma_matriculados=np.sum(inscritos_1)-np.sum(inscritos_2)

print(suma_matriculados)
```

```python
copia=np.sort(copia)
n_iteraciones=10000
diffs=np.zeros(n_iteraciones)
for i in range(n_iteraciones):
    np.random.shuffle(copia)
    inscritos_fake_1 = copia[:n1]
    inscritos_fake_2 = copia[n2:]
    diffs[i] = np.sum(inscritos_fake_1-inscritos_fake_2)
```

```python
plt.vlines(suma_matriculados, 0,700,color='r')
_ = plt.hist(diffs,bins=40)
```

```python
np.count_nonzero(diffs<suma_matriculados)/len(diffs)
```
